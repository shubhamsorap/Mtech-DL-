```python
#h and k
int_cols_toimpute = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm']
auto[int_cols_toimpute] = auto[int_cols_toimpute].astype('float64')
auto[int_cols_toimpute] = auto[int_cols_toimpute].fillna(auto[int_cols_toimpute].mean())
auto[int_cols_toimpute].isnull().sum()
```

```python
#i 
auto['num-of-doors'].fillna(auto['num-of-doors'].mode()[0], inplace=True)
auto.isnull().sum()
```

```python
#j. reset the index
auto.reset_index(inplace = True) 
```

```python
#l
auto['city-L/100km'] = 235/auto['city-mpg']
auto['highway-L/100km'] = 235/auto['highway-mpg']
auto.head()
```

```python
#m
auto['length'] = auto['length']/auto['length'].max()
auto['width'] = auto['width']/auto['width'].max()
auto['height'] = auto['height']/auto['height'].max()
```
